```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.dates as mdates
from datetime import datetime
from scipy.stats import kstest
from scipy.stats import ks_2samp
from pandas.core.frame import DataFrame
```

```python
#input data
def read_csv_data():
    music_data = pd.read_csv("/Users/cynthiadu/Downloads/top200_weekly_US_2016_2019.csv", header=None, engine='python')#imput music data
    df = pd.DataFrame(music_data)
    X_data = df.loc[0:25000:,]
    y_data = df.loc[:,0:2]
    return X_data, y_data, df
```

```python
x_data, y_data, df = read_csv_data()
df = df.drop(0 , axis = 0)
artist_stream = y_data.loc[:,[0, 2]]

#create dataframe for weeks
weeka = df.iloc[:,3]
weekb = weeka.tolist()
week = sorted(set(weekb), key = weekb.index)
week = DataFrame(week)
```

```python
#extract different musics 
music = df.values[:,1]
music_multi = set(music)
music_single = [l for l in music_multi]
```

```python
for k in music_single : 
    #distribution test for music
    music_array = df.loc[df[1] == k, 1:3]
    music_array1 = df.loc[df[1] == k, 2]
    music_arraylist = list(map(int, music_array1.values))
    test_stat_music = kstest(music_arraylist, 'norm')
    print(test_stat_music)
    
    #distribution plot of music

    allweek_music = pd.concat([week, music_array.iloc[:,1]], axis = 1)
    allweek_music = allweek_music.fillna(0)
    allweek_music.set_index(allweek_music.iloc[:,0])
    allweek_music = allweek_music.iloc[:,1].astype(float)
    allweek_music.plot(figsize=(8, 4))
    
    plt.ylabel(u'streams')
    plt.xlabel(u'week')
    plt.legend(loc="best")
    plt.grid(True)
    plt.savefig('/Users/cynthiadu/test.png')
    fig = plt.gcf()
    plt.show()
```

```python
ra = []
for i in music_single : 
    #relative change of music
    music_array = df.loc[df[1] == i, 1:3]
    music_array1 = df.loc[df[1] == i, 2]
    music_arraylist = list(map(int, music_array1.values))
    for j in range(len(music_arraylist)-1) :
        rc = (music_arraylist[j+1] - music_arraylist[j]) / music_arraylist[j]
        ra.append(rc)
        #plot of relative change of music
    #ra.plot(kind='hist', density=True, alpha=0.3, label='Relative Change', color='red', loglog=True, bins=np.logspace(np.log10(min(ra)),np.log10(max(ra)), 50)
    
        fig1 = plt.figure(1)
        ax1 = fig1.add_subplot(1,1,1)
        ax1.plot(ra, color = 'red', label = 'i') 
        ax1.set_xticklabels(week)
        ax1.legend(loc="best")
        plt.show()
        #distribution test of the relative change of music 
        ra_kstest = kstest(ra, 'norm')
        print(ra_kstest)
```
